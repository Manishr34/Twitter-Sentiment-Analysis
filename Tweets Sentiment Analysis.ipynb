### Cleaning Tweets

```python
import nltk
from nltk.corpus import stopwords
import string
def text_cleaning(text):
    rem_punch = [char for char in text if char not in string.punctuation]
    rem_punch_join = "".join(rem_punch)
    clean_msg = [word for word in rem_punch_join.split() if word.lower() not in stopwords.words("english")]
    return clean_msg
```

```python
clean_tweets = df["Text"].apply(text_cleaning)
```

### Vectorizing Tweets

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer=text_cleaning)
tweets_vectorizer = CountVectorizer(analyzer = text_cleaning,dtype = 'uint8').fit_transform(df["Text"]).toarray()
```

```python
tweets_vectorizer.shape
```

```python
X = tweets_vectorizer
y = df["Sentiment"]
```

### Creating Naive Bayes Classification Model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
y_pred = NB_classifier.predict(X_test)
cm =confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm,annot=True)
```

###  Creating Report

```python
classification_report(y_test,y_pred)
```
